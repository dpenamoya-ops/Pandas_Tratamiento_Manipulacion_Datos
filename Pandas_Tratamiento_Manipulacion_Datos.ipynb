##2.Datos numéricos

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[3:])
datos
```

```python
datos.reset_index(inplace=True, drop=True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
#solo muestra información
datos['max_hospedes'].astype(np.int64)
```

```python
#transforma el registro
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
```

```python
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
datos['precio'.astype(np.float64)]
#el error AttributeError ocurre debido a que existe un simbolo $ antes del número
```

```python
#remueve "$", "," y espacios
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos.info()
#registro precio sigue siendo tipo object ya que no hemos modificado y¡su tipo (sólo eliminamos caracteres extra)
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
datos[['cuota_deposito', 'cuota_limpieza']].apply(lambda x: x.replace('$', '').replace(',', '').strip())
#no funciona ya que apply solo funciona para una columna
```

```python
#para varias columnas debemos usar applyma
datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']].applymap(type)
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

##3.Datos de texto

```python
datos.head()
```

```python
datos['descripcion_local']  = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
datos['descripcion_local'][3169]
```

```python
#datos['descripcion_local'].str.replace('PALABRA A BUSCAR', ' ', regex=True)
datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
#\- ❌ NO es una secuencia de escape válida
#¿Qué hace r""? Le dice a Python: “No interpretes los \, pásalos tal cual a la expresión regular”
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r"[^a-zA-Z0-9\-']", " ", regex=True)
```

```python
datos['descripcion_local']
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r'(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
datos['descripcion_local']
```

```python
datos['descripcion_local'].head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
datos.head()
```

```python
datos['comodidades'] = datos['comodidades'].str.replace(r'\{|}|\"', '', regex=True)
```

```python
datos.head()
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
```
